社会科学一直尝试对事件之间的因果性进行回答。因果性就像是伊甸园里的禁果：尽管所有人都告诫你不要去尝试回答因果性，做一只把头埋在沙子里的鸵鸟，只说明统计相关性去逃避因果的追寻，但是，你总是试图去摘取它，因为回答因果性实在有着致命的诱惑。同样是七月，Pearl的*The Book of Why*出了中文版，取名*《为什么》*。这名字实在是糟，我觉得作者要是看到这么俗气的译名，也会问一句：为什么？！本书作者是大名鼎鼎的Causality一书的作者，主要研究计算机领域的因果推断。他在书中倡导贝叶斯方法在因果推断的运用，并且认为因果推断的前提是要有因果假设，而不能直接由数据得出因果关系，也就是说，即使是同一数据集，也会因为假设不同而得出不同的结论，而这一假设关系由因果图来表示。

譬如说，这里有两张表，一张是不同手术方法对肾结石的疗效，一张是不同药物的降压效果。注意，这里的数据是一致的，但是它们有不同的因果图，决定所得结论是否合理。我们单看数据本身，可以看出不论是哪个分组，A的疗效都要比B好，但是一旦取均值后，结论就颠倒过来，这一现象被称为辛普森悖论。也就是基于(1)(2)不能推得(3)。


\begin{align}
\frac{a_1}{b_1}&>\frac{c_1}{d_1} \quad(1)\\
\frac{a_2}{b_2} &> \frac{c_2}{d_2}\quad(2)\\
\frac{a_1+a_2}{b_1+b_2}&>\frac{c_1+d_1}{d_1+d_2}\quad(3)
\end{align}

我们可以通过右侧的因果图来看各个成分之间的关联。其中，对于上面的表格中的两列来说，肾结石的大小影响疗法的选择，而治疗影响结果，因此将这两个变量作平均处理没有什么实际意义。但是对于下面的表格来说，不论血压是高还是低，其最终目的都是将血压恢复正常值，因此取均值有意义，这时可以说，综合来看，疗法B对于恢复正常血压更有效。
这样我们印证了先前的命题，即单一的数据集本身不能得出因果性结论，必须要先有合理的假设才行，这一假设的提出和社会科学的理论密切相关。

进一步说，辛普森悖论是否得以成为一个悖论建立在组间的异质性是否对假设有密切关系。由于社会科学常常处理分类变量，要想避免辛普森带来的混淆，就要想办法得出分层效应，找个各个效应对应的子群体(subgroups)，才不会因为平均化而使得结论歪曲或者根本一无所获。当然，这个话题已经不在现在的讨论范围里面了，最后直观地解释了辛普森悖论：两个回归系数$\beta$为负的子群体合并后再做一次OLS，其回归系数却为正。

## 2. 机器学习的预备知识


广义上机器学习有两种思路，前者是从数据中提取特征，然后进行拟合以适用新的数据，来完成预定的任务。譬如想要预测房价，则可以抽取房间数、面积、地理位置等特征来作为变量送入模型进入训练；后者通过多层网络(layers)来循序渐进地表示某些难以抽离出特征，譬如要想识别出图片中的猫，则先训练一层识别线条的网络、再是边和角、最后是猫。后者因为这种层层递进的“深”的特征得名深度学习。本文仅仅涉及前者，事实上社会科学在运用这些算法时需要相当谨慎地选取特征（变量）和模型，这对于文章的解释力十分关键。

机器学习的基础任务是使用模型来拟合(fit)数据，并且在新的数据集上这一模型仍然能有较高的适用性(generalization)。如果你过于追求贴合数据，这时就常常会发生过拟合(overfitting)的问题，因此一个最佳的模型要在训练误差 E_{train}和验证误差 E_{test}之间取得平衡。E_{train}是社科中常常出现的指标，以最常见的线性回归（OLS）为例，我们熟知的R方就是1-E_{train}。

相比之下，E_{test}大家可能就很少听说了。一般而说，在没有新的数据集可以验证的情况下，我们会对样本数据进行拆分，一部分作为训练集，一部分作为测试集。在社会科学中，因为样本数据宝贵，禁不起拆分，我们会把所有得到的样本投入模型，而不考虑模型检验的问题。不过，对于决策树和随即森林来说，由于不使用拆分方法，而通过限制树的深度、叶子数量等等来进行regularization。当然，也可以不使用拆分的方法让原来就很有限的样本损失，这个讨论EM算法的时候再说。

###  2.1 机器学习的一般范式

ERM(Empirical Risk Minimization) 是机器学习的一个基本的范式，我们可以用这个框架来囊括本文下面要讨论的所有模型。

对ERM的几个部分作说明：

$arg_\theta min$指后面的等式取最小值时变量$x$在参数为$\theta$时的取值

$N$意味着模型执行了$N$次，取每次的$arg min$的均值

$l$ 为损失函数，$f$为训练模型，$t$ 为该模型共有$t$个变量被放入。如果$f$是OLS，那么这里的损失函数就是最小二乘，即最小化残差的平方，如果是逻辑斯蒂回归，那么这里的损失函数就是MLE，也就是最大似然概率。当然虽然它叫“最大”，但是取值是仍是导数的最小值，只不过此时取值的$y$为最大概率。

$\lambda \Omega (\theta)$这个部分很有意思，它是一个用于Regularization的算子，

其目的是施加一定的惩罚来选择一个不那么优秀的模型来避免过拟合的可能性$

\begin{equation}
arg_\theta min \frac{1}{N} \sum_n l(f(x^{(t)}; \theta), y^{(t)}) + \lambda \Omega (\theta)
\end{equation}
